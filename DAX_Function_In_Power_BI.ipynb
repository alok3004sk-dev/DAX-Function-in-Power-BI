Q4 :- Using DAX, create a measure that calculates cumulative Year-to-Date (YTD) Sales based on the Order
Date
- Answer

Step 1: Open Power BI

Step 2: Select the Table

Step 3: Create a New Measure

Step 4: Enter the DAX Formula = YTD = TOTALYTD(SUM(AmazonSales[SalesAmount]),AmazonSales[OrderDate])

Step 5: Press Enter.

Q6 :- Create a calculated column classifying orders as “High Value” if sales > 2000, else “Low Value”.

- Answer  

Step 1: Open Power BI

Step 2: Go to Data View

Step 3: Create a New Column name OrderCategory

Step 4: Enter the DAX Formula
OrderCategory = IF(AmazonSales[SalesAmount] > 2000 , "High Value" ,"Low Value")

Q7 :- Create a measure to calculate total sales from the Online channel only.

- Answer

Step 1: Open Power BI

Step 2: Create a New Measure

Step 3: Enter the DAX Formula
OnlineSales = CALCULATE(SUM(AmazonSales[SalesAmount]),AmazonSales[SalesChannel]="Online")

Q8 :- Create a measure called "Date Difference" that calculates the number of days between the latest  Order
Date and today's date using the DATEDIFF function.

- Anwer

Step 1: Open Power BI

Step 2: Create a New Measure

Step 3: Enter the DAX Formula
DateDifference = DATEIF(MAX(AmazonSales[OrderDate]),TODAY(),DAY)
